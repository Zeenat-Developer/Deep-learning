## Improve build_model function

### Subtask:
Modify the `build_model` function to allow Keras Tuner to search for the optimal number of hidden layers and the number of units in those layers, in addition to the optimizer. This will involve using `hp.Int` for these new hyperparameters.


**Reasoning**:
The subtask requires modifying the `build_model` function to allow Keras Tuner to search for the optimal number of hidden layers and units per layer, in addition to the optimizer. This involves iterating through a hyperparameter-defined number of layers and dynamically setting the number of units for each layer.



```python
def build_model(hp):
    model=Sequential()

    num_layers = hp.Int('num_layers', min_value=1, max_value=3, step=1)

    for i in range(num_layers):
        if i == 0:
            model.add(Dense(hp.Int(f'num_units_{i}', min_value=16, max_value=128, step=16), activation='relu', input_dim=8))
        else:
            model.add(Dense(hp.Int(f'num_units_{i}', min_value=16, max_value=128, step=16), activation='relu'))

    model.add(Dense(1,activation="sigmoid"))
    optimizer = hp.Choice('optimizer',values=['sgd', 'adam', 'rmsprop', 'adadelta'])



    model.compile(loss="binary_crossentropy",optimizer=optimizer,metrics=["accuracy"])
    return model
```

## Run Keras Tuner with improved build_model

Initialize and run `kt.RandomSearch` with the updated `build_model` function to find the best combination of hyperparameters, including the number of layers and units.


**Reasoning**:

```python
tuner = kt.RandomSearch(build_model, objective='val_accuracy', max_trials=10, directory='my_dir', project_name='my_model_tuning')
tuner.search(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
tuner.results_summary()
```

**Reasoning**:

```python
best_model = tuner.get_best_models(num_models=1)[0]
print("Retrieved best model from Keras Tuner.")
```

**Reasoning**:

```python
history = best_model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
print("Best model trained for 100 epochs.")

loss, accuracy = best_model.evaluate(X_test, y_test)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```
